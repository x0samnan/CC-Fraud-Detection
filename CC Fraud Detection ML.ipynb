Importing dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading Credit card dataset to Pandas DataFrame

```python
cc_data = pd.read_csv('creditcard.csv')
```

Dataset information

```python
cc_data.info()
```

First five rows of dataset

```python
cc_data.head()
```

Last five rows of dataset

```python
cc_data.tail()
```

Checking column wise null values

```python
cc_data.isnull().sum()
```

No missing values found

Distribution of Fraudlent Transaction and Legit Transaction


```python
cc_data["Class"].value_counts()
```

0 -> Legit Transaction
1 -> Fraudlent Transaction

Since, the data is unbalanced, we will handle the data by separating and analyzing it

```python
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Comparing values of both transactions

```python
cc_data.groupby('Class').mean()
```

Undersampling <br />
Building a sample dataset <br />
Fraudlent transactions -> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenation of Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

Checking the initial and final values of new Dataframe

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

Re-checking the value counts

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting of data into features and targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class'] 
```

```python
print(X) 
```

```python
print(Y)
```

Splitting data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Logistic Regression

```python
model = LogisticRegression(max_iter=3000) 
```

Training logistic regression model with training data

```python
model.fit(X_train, Y_train)
```

Accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

Accuracy on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```
